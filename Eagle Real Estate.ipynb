# In this project, i am creating a 2-dimesional array for a real estate company.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# creating the Data as python dictionary
eagle_Estate = {
    'name': ['John', 'Ashely', 'Daniel', 'Bishop', 'Andrew', 'Melissa'],
    'Property_Type': ['House', 'Family Unit', 'Condo', 'Family Unit', 'Townhome', 'House'],
    'bed':[4, 5, 3, 5, 4, 3],
    'bath': [3, 4, 2, 4, 3, 2],
    'Location': ['Fishers', 'Greenwood', 'Avon', 'Fishers', 'Greenwood', 'Avon'],
    'price_sold': np.random.randint(350000, 500000, 6), 
    'date_sold': pd.date_range('1/1/2022' , '12/12/22', 6)
}
```

```python
# Generate a customized index as a unique employee id. The stop Integer does not count in the array
empId = np.arange(201, 207)
```

```python
# Combining our dataSets and Index as row and columns
Sale_Table = pd.DataFrame(eagle_Estate, index=empId)
Sale_Table
```

# PREVIOUS YEAR SALES

```python
# lets assume that the same agents sold a house in 2021 and create table for that.
Previous_Sale = {
    'name': ['John', 'Ashely', 'Daniel', 'Bishop', 'Andrew', 'Melissa'],
    'Property_Type': ['House', 'Family Unit', 'Condo', 'Family Unit', 'Townhome', 'House'],
    'bed':[5, 4, 3, 6, 4, 5],
    'bath': [3, 3, 5, 4, 3, 3],
    'Location': ['Beech Grove', 'Lebanon', 'Anderson', 'Fishers', 'Fort Wayne', 'Avon'],
    'price_sold': np.random.randint(500000, 900000, 6), 
    'date_sold': pd.date_range('1/1/2021' , '12/12/21', 6)
}

# Generate a customized index as a unique employee id. The stop Integer does not count in the array
empId2 = np.arange(201, 207)

# converting the previous sale table to a DataSet
Previous_Sale_Table = pd.DataFrame(Previous_Sale, index= empId2)
Previous_Sale_Table
```

# APPENDING BOTH THE SALE_TABLE AND PREVIOUS_SALE_TABLE

```python
Total_Sales_Table = Sale_Table.append(Previous_Sale_Table)
Total_Sales_Table
```

```python
# Sorting from smallest to largest price
Total_Sales_Table.sort_values('price_sold')
```

# DATA EXPLORATION AND ANALYSIS


```python
# finding out the average price of houses in all locations
Total_Sales_Table.loc[:, 'price_sold'].mean()
```
